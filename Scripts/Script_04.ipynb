```python
import urllib.request

text_file = open('../Data/summit_post_urls_selected.txt', "r")
lines = text_file.read().split("\n")
text_file.close()
```

```python
lines
```

```python
from os.path import join, exists
from os import listdir, makedirs

if not exists('../Data/helmet_images'):
        makedirs('../Data/helmet_images')
```

```python
nbfiles = len(lines)
for url in range(nbfiles) :
    with urllib.request.urlopen(lines[url]) as response, open('../Data/helmet_images/'+str(url)+'.jpg', 'wb') as out_file:
            data = response.read()
            out_file.write(data)
```
